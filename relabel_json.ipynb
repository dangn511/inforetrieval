```python
with open('cleaned_jd_3.json', 'r') as fp:
    master_list = json.load(fp)
    
relabel = {}

for idx, item in enumerate(master_list):
    print(idx, master_list[item]["job_id"])
    relabel[idx] = master_list[item]
    
with open('master_list_relabeled.json', 'w') as fp:
    json.dump(relabel, fp)
    print('saved')
```
